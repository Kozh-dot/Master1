## Algorithme de descente du gradient

Code des fonctions et tests de ces fonctions dans le cas de la régression simple et multiple

ATTENTION : les paramètres des fonctions sont à compléter


```python
#-- Modèle
def f(X,betas):
    return np.dot(X,betas)

Yhat1 = f(X1,[0,1])
#print(Yhat1.shape)
print(Yhat1[:5])

#-- Test 1 - reg simple


#-- Test 2 - reg multiple
Yhat3 = f(X3,[0,1,-1,2])
#print(Yhat3.shape)
print(Yhat3[:5])
```

```python
#%% Fonction-cout
def cout(X,Y,betas):
    m,d = X.shape
    return (1/(2*m))*sum((f(X,betas)-Y)**2)
        

#%% Test 1 - reg simple
Ycout1 = cout(X1,Y,[0,1])
print(Ycout1)
#%% Test 2 - reg multiple
Ycout3 = cout(X3,Y,[0,1,-1,2])
print(Ycout3)
```

```python
#%% Gradient
def grad(X,Y,betas):
    m,d = X.shape
    return 1/m * np.dot(f(X,betas)-Y,X)

#%% Test 1 - reg simple
print(grad(X1,Y,[0,1]))
#%% Test 2 - reg multiple
print(grad(X3,Y,[0,1,-1,2]))
```

```python

#--- standardisation des X (données centrées-réduites)
# attention : ne pas standardiser X^0...
def standardisation(X):
    Xtmp = np.divide(np.subtract(X,np.mean(X,axis=0)),np.std(X))
    Xtmp[:,0] = 1
    return Xtmp
    
#-- Test 1 - reg simple
print(X1)
print(standardisation(X1))

#-- Test 2 - reg multiple
print(X3)
print(standardisation(X3))
```

```python
#%% Algo de descente du gradient
def grad_descent(X,Y,betas,pas):
    X = standardisation(X)
    new_beta = betas + pas * grad(X,Y,betas)
    err = cout(X,Y,new_beta)
    i=0
    while i<1000 and err > 0.01:
        
        new_beta = new_beta - pas * grad(X,Y,new_beta)
        err = abs(err- cout(X,Y,new_beta))
        print(err)
        i+=1
    
        
#-- Test 1  - reg simple
grad_descent(X1,Y,[-1,1],0.000001)

#-- Test 2 - reg multiple
```

```python
#-- Fonction-cout
# ordre des arguments : beta en 1er pour comparer avec la fonction d'optim
def cout(X,Y,betas):
    m,d = X.shape
    return (-1/(m)) * np.sum(np.sum(Y, np.log(f(X,betas))), np.sum(1 - Y, np.log(1 - f(X,betas))))

#-- Test 1 - reg simple

#-- Test 2 - reg multiple
```

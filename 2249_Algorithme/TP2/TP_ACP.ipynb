```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D  
```

```python
def affiche3D(df,xlim=0, ylim = 0, zlim = 0):
    fig = plt.figure(figsize=(8,6))
    ax = Axes3D(fig)

    dx = df['Intensite odeur']
    dy = df['Glucose']
    dz = df['Acide citrique']

    ax.scatter(dx,dy,dz, s=50)
    
    for i,p in enumerate(data.index):
        ax.text(dx[i]+0.01, dy[i]+0.01, dz[i]+0.01, p,fontsize=12)

    ax.set_xlabel('Intensite odeur')
    ax.set_xlabel('Glucose')
    ax.set_xlabel('Acide citrique')
    if(xlim != 0):
        ax.set_xlim(-xlim,xlim)
    if(ylim != 0):
        ax.set_ylim(-ylim,ylim)
    if(zlim !=0):
        ax.set_zlim(-zlim,zlim)
    plt.show
    
affiche3D(data)
```

* Centrer l'ensemble des donn√©es (soustraire la moyenne) (cr√©er une variable ``Y``)
* Revisualiser les donn√©es

```python
Y = data - data.mean(axis=0)
affiche3D(Y)
```

* Centrer et r√©duire l'ensemble des donn√©es (soustraire la moyenne et diviser par l'√©cart-type) (cr√©er une variable ``Z``)
* Revisualiser les donn√©es

```python
Z = Y / data.std(ddof=0)
```

```python
affiche3D(data)
affiche3D(Y)
affiche3D(Z)
```
